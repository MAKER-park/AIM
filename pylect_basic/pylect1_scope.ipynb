```python
a = 5           # a is assigned in the interpreter, so it's global
def func1(b):    # b is in the function's argument list, so it's local           
  c = a + b     # c is only assigned inside the function, so it's local
  return c  

print (func1(4)) # gives 4+5=9
#print (c)       # Error! not defined
print (a) # gives 5
```

```python
def func2(b):
    global c
    c = a + b
    return c
```

```python
print (func2(4)) # gives 4+5=9
print (c)       # not it is defined 9
print (a) # gives 5
```

```python
a = 5           # a is assigned in the interpreter, so it's global
def func1(b):    # b is in the function's argument list, so it's local           
  c = a + b     # c is only assigned inside the function, so it's local
  return c

print (func1(4)) # gives 4+5=9
print (c)       # Error! not defined
print (a) # gives 5
```

```python
a = 5           # global
def func2(b):  
    global c
    c = a + b
    return c
print (func2(4)) # gives 4+5=9
print (c)       # now it was defined as 9
print (a) # gives 5
```

```python
a = 5           # global
def func2(b):  
    global c
    c = a + b
    a=a+1
    return c
print (func2(4)) # gives 4+5=9
print (c)       # now it was defined as 9
print (a) # gives 5
```

Scope for immutable and mutables

```python
def func5(x,y):
    x = x * 2
    y = y * 2
    print (x, y)  
    
def func6(x,y):
    x = x * 2
    y[0] = y[0] * 2
    print (x, y)  
```

```python
a = 2
b = [1,2]
func5(a,b)
print (a, b)      # 0 [1, 2]
```

```python
func6(a,b)
print (a, b)      # 0 [2, 2]
```
